# ENCODEING

### 1. Lable Encoding (yes/no)
### 2. One-hot-encoding (two or multiple)

```python
dataset['sex'].value_counts()
```

```python
dataset['sex'] = dataset['sex'].map({"male" : 0,"female" : 1})
```

```python
dataset
```

```python
# same  as smoker
dataset['smoker'].value_counts()
```

```python
dataset['smoker'] = dataset['smoker'].map({"no" : 0,"yes" : 1})
```

```python
# Rename sex-> isfemale and smokerr -> issmoker
dataset.rename(columns={
    'sex' :'is_female',
    'smoker': 'is_smoker'
},inplace = True)
```

```python
dataset.head()
```

```python
# One-hot-encoding
dataset = pd.get_dummies(dataset,columns = ['region'],drop_first=True)
```

```python
dataset.sample(5)
```

```python
dataset = dataset.astype(int)
```

```python
dataset.head()
```

# Feature Scaling

```python
# bmi divided into 4 type
dataset['bmi_category'] = pd.cut(
    dataset['bmi'],
    bins=[0, 18.5, 24.9, 29.9, float('inf')],
    labels=['Underweight', 'Normal', 'Overweight', 'Obese']
)
```

```python
dataset
```

```python
# one-hot-encoding
dataset = pd.get_dummies(dataset,columns = ['bmi_category'],drop_first=True)
```

```python
dataset = dataset.astype(int)
```

```python
dataset.head()
```

# Use Standard Scaler for renge -1 to 1

```python
# 'age','bmi','children'  Standardscaler using in -1 to 1 conver  value
from sklearn.preprocessing import StandardScaler
cols = ['age','bmi','children']
scaler = StandardScaler()

dataset[cols] = scaler.fit_transform(dataset[cols])
```

```python
dataset.head()
```

# Separate features and target

```python
X = dataset.drop(columns='charges', axis=1)
y = dataset['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split data train and test 
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=2
)
```

# Train a regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# predict 
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
train_pred 
```

test_pred

# : Evaluate the model

```python
from sklearn import metrics
```

### R² Score

```python
print(metrics.r2_score(y_train, train_pred))
print(metrics.r2_score(y_test, test_pred))
```

### Mean Absolute Error

```python
print(metrics.mean_absolute_error(y_test, test_pred))
```

### Mean Squared Error

```python
print(metrics.mean_squared_error(y_test, test_pred))
```

### RMSE

```python
print(np.sqrt(metrics.mean_squared_error(y_test, test_pred)))
```

# Using all model Regression

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor

from sklearn.metrics import r2_score
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(),
    "Lasso": Lasso(),
    "Elastic Net": ElasticNet(),
    "Decision Tree": DecisionTreeRegressor(random_state=2),
    "Random Forest": RandomForestRegressor(random_state=2),
    "SVR": SVR(),
    "XGBoost": XGBRegressor(random_state=2)
}

for name, model in models.items():
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    score = r2_score(y_test, prediction)
    print(f"{name}: {score:.4f}")
```
