# Encode the data

# encoding Fule_Type

```python
car_dataset.replace({
    'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}
},inplace = True)
```

### Seller_Type encoding

```python
car_dataset.replace({
    'Seller_Type':{'Dealer':0,'Individual':1}},inplace = True
)
```

### Transmission encoding

```python
car_dataset.replace({
    'Transmission':{'Manual':0,'Automatic':1}},inplace = True
)
```

```python
car_dataset.sample(10)
```

# Spliting data into train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X = car_dataset.drop(['Car_Name','Selling_Price'], axis = 1)
y = car_dataset['Selling_Price']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Apply Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
linear = LinearRegression()
```

```python
linear.fit(X_train,y_train)
```

# Model Evaluation

# Prediction on  Training data

```python
pred_train = linear.predict(X_train)
```

```python
pred_train
```

# R Squared error

```python
erro_score = metrics.r2_score(y_train, pred_train)
print(f'R squared error is :{erro_score}')
```

# Visualize the ctual prices and Predicted prices

```python
sns.scatterplot(x = y_train,y = pred_train)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual price vs predicted price")
plt.show()
```

# Prediction on Training data 

```python
pred_test= linear.predict(X_test)
```

```python
pred_test
```

```python
erro_score_test = metrics.r2_score(y_test, pred_test)
print(f'R squared error is :{erro_score_test}')
```
